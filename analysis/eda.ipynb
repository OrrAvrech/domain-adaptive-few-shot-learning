```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = Path("/Users/orrav/Documents/Data/domain-adaptive-few-shot-learning/filtered_annotations.csv")
df = pd.read_csv(csv_path)
df.head()
```

```python
def plot_gb(group, x_label = "Class", y_label = "Mean"):
    group.plot(kind='bar')  # You can choose different plot types based on your data and preference
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```

```python
cls_counts = df["final_class"].value_counts()
plot_gb(cls_counts)
```

```python
gb_rating = df.groupby(by="final_class")[["abstraction", "clarity"]].mean()
plot_gb(gb_rating, x_label="Class", y_label="Mean (1-5)")
```

```python
gb_rating_annotator = df.groupby(by=["annotator"])[["abstraction", "clarity"]].mean()
plot_gb(gb_rating_annotator, x_label="Annotator ID", y_label="Mean (1-5)")
```

```python
gb_class_annotator = df.groupby(by=["annotator"])["final_class"].value_counts()
plot_gb(gb_class_annotator, x_label="Annotator ID, Class", y_label="Counts")
```
